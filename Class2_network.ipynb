# Class 2 network example

```python
def create_class2_network():
    G = nx.Graph()
    
    for i in range(10):
        G.add_node(i)
    
    edges_to_add = [
        (0,1,0.98), (0,6,0.96), (2,3,0.95), (9,7,0.92),

        (1,5, 0.89), (6,5, 0.82), (8,4,0.87), (8,14,0.89), (14,16,0.85), (17,16,0.88), (17,15, 0.82), (15,13,0.86), (13,4,0.82),

        (6,9,0.73), (3,8,0.77), (6,18,0.74), (18,19,0.73), (19,10,0.71), (10,11, 0.76),

        (5,7,0.69), (2,8,0.63), (11,12,0.68), (13,12,0.66), (2,1,0.65),

        (6,1,0.55), (9,16,0.58), (15, 7, 0.59), 

        (7,8,0.45), (10,0, 0.47), (11,14,0.42),

        (5,2,0.39), (19,3,0.35),

        (9,5, 0.27), (8,3,0.24)
    ]

    for u, v, weight in edges_to_add:
        if u not in G:
            G.add_node(u)
        if v not in G:
            G.add_node(v)
        G.add_edge(u, v, weight=weight)

    pos = nx.spring_layout(G, seed=42, k=0.3, iterations=100)
    return G, pos

# Crea la rete
G2, fixed_pos2 = create_class2_network()
title2 = "Class 2 Network"
analyze_persistence(G2, fixed_pos2, title2)
```

```python
diagrams2 = get_persistence_diagram(G2, title2, print_diagram=False)  
h12, h1_tilde2 = compute_hollowness(G2, diagrams2, homology_dim=1)
print("Hollowness h1:", h12)
print("Chain-length normalized hollowness ~h1:", h1_tilde2)
```

Null models

```python
G_sh2, pos_sh2=randomize_weights(G2, fixed_pos2)
title_sh2 = "with Randomized Weights"
analyze_persistence(G_sh2, pos_sh2, title_sh2)
diagrams_sh2 = get_persistence_diagram(G_sh2, title_sh2, print_diagram=False)  
h1_sh2, h1_tilde_sh2 = compute_hollowness(G_sh2, diagrams_sh2, homology_dim=1)
print("Hollowness h1:", h1_sh2)
print("Chain-length normalized hollowness ~h1:", h1_tilde_sh2)
```

```python
G_rnd2, pos_rnd2=randomize_edges_and_weights(G2, fixed_pos2)
title_rnd2 = "with Randomized Edges"
analyze_persistence(G_rnd2, pos_rnd2, title_rnd2)
diagrams_rnd2 = get_persistence_diagram(G_rnd2, title_rnd2, print_diagram=False)  
h1_rnd2, h1_tilde_rnd2 = compute_hollowness(G_rnd2, diagrams_rnd2, homology_dim=1)
print("Hollowness h1:", h1_rnd2)
print("Chain-length normalized hollowness ~h1:", h1_tilde_rnd2)
```
