# Class 1 network example

```python
def create_class1_network():
    G = nx.Graph()
    
    for i in range(20):
        G.add_node(i)

    edges_to_add = [
        (0,1,0.98), (0,6,0.96), (2,3,0.95), (9,7,0.92),

        (1,5, 0.89), (6,5, 0.82), (8,4,0.87), (10, 11, 0.88), (2, 10, 0.85), (3,12, 0.84), (12,11, 0.81), (4, 15, 0.89), (15,17, 0.88), (17,16, 0.86), (16,14, 0.85), (14,8,0.83), (19,0,0.83), (19,18, 0.81), (18,6,0.84),

        (6,9,0.73), (3,8,0.77),

        (5,7,0.69), (2,8,0.63), (18,0,0.66), (14,4,0.65), (14,15, 0.67), (16,15,0.61),

        (6,1,0.55), (9,14,0.58), (13,15,0.52), (13,12,0.59),

        (7,8,0.45), (2,3,0.47), (2,11,0.42), (3,12, 0.44), 

        (5,2,0.39), (9,8,0.33), (4,11,0.31), (4,12, 0.37), (1,13, 0.34),

        (9,5, 0.27), (8,3,0.24), (3,9,0.26),(9,2,0.22), (2,8,0.28), (2,4,0.23),

        (7,2,0.11)
    ]


    for u, v, weight in edges_to_add:
        if u not in G:
            G.add_node(u)
        if v not in G:
            G.add_node(v)
        G.add_edge(u, v, weight=weight)

    pos = nx.spring_layout(G, seed=42, k=0.3, iterations=100)
    return G, pos


G, fixed_pos = create_class1_network()
title = "Class 1 Network"
analyze_persistence(G, fixed_pos, title)
```

```python
diagrams = get_persistence_diagram(G, title, print_diagram=False)  
h1, h1_tilde = compute_hollowness(G, diagrams, homology_dim=1)
print("Hollowness h1:", h1)
print("Chain-length normalized hollowness ~h1:", h1_tilde)
```

Null models

```python
G_sh, pos_sh=randomize_weights(G, fixed_pos)
title_sh = "Class 1 Network with Randomized Weights"
analyze_persistence(G_sh, pos_sh, title_sh)
diagrams_sh = get_persistence_diagram(G_sh, title_sh, print_diagram=False)  
h1_sh, h1_tilde_sh = compute_hollowness(G_sh, diagrams_sh, homology_dim=1)
print("Hollowness h1:", h1_sh)
print("Chain-length normalized hollowness ~h1:", h1_tilde_sh)
```

```python
G_rnd, pos_rnd=randomize_edges_and_weights(G, fixed_pos)
title_rnd = "Class 1 Networkwith Randomized Edges"
analyze_persistence(G_rnd, pos_rnd, title_rnd)
diagrams_rnd = get_persistence_diagram(G_rnd, title_rnd, print_diagram=False)  
h1_rnd, h1_tilde_rnd = compute_hollowness(G_rnd, diagrams_rnd, homology_dim=1)
print("Hollowness h1:", h1_rnd)
print("Chain-length normalized hollowness ~h1:", h1_tilde_rnd)
```
